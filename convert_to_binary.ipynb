```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
# Enter the filename and extension of the input image
file_extension = ".png"
filename = input("Enter the filename: ")
```

```python
# Read the image, convert it to grayscale, and display it and its pixel values

# Contruct the path to the 'annotations' subfolder in the current working directory
directory = os.path.join(os.getcwd(), 'annotations')
file_path = os.path.join(directory, filename + file_extension)

# Read the image
img = cv2.imread(file_path)

if img is not None:
    # Convert image to grayscale
    img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # Display the image
    plt.imshow(img, cmap='gray')
    plt.title('Original Image')
    plt.axis('off')
    plt.show()

    # Get and print unique pixel values
    original_pixels = np.unique(img)
    print("Pixel Values:")
    for value in original_pixels:
        print(value)

else:
    print(f"Error: Unable to load image from path {file_path}")
```

```python
# Checks each pixel to see if the value is greater than 0
# If the condition is true, sets its value to 1.0. Otherwise, 0.0.
binary_img = np.where(img > 0, 1.0, 0.0)

# Display the image
plt.imshow(binary_img, cmap='gray')
plt.title('Binary Image')
plt.axis('off')
plt.show()

# Check the conversion by printing the unique pixel values
binary_pixels = np.unique(binary_img)
print("Pixel Values:")
for value in binary_pixels:
    print(value)
```

```python
from skimage.io import imsave
```

```python
# Save the binary image to a specified directory

# Construct the path to the 'label' subfolder in the current working directory
output_directory = os.path.join(os.getcwd(), 'label')

# Saves the binary image as a .png with the same filename as the input file
output_file_path = os.path.join(output_directory, filename + ".png")
binary_uint8 = binary_img.astype('uint8')
imsave(output_file_path, binary_uint8)

print(f"Image saved as '{output_file_path}'")
```
